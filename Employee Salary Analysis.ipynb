```python
# Sample data
labels = ['M', 'F']

# Plotting the pie chart
plt.figure(figsize=(8, 8))  # Set the size of the plot
plt.pie(gp, labels=labels, autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle

# Adding a title
plt.title('Gender Proportion')

# Display the plot
plt.show()
```

```python
#The average salaries for each gender in different Departments 
avgG= df.groupby(["Department_Name", "Gender"])['Base_Salary'].mean().reset_index(name='Base_Salary').set_index("Department_Name")
avgG.head()
```

```python
#Males in each department
males = avgG[avgG["Gender"]== 'M']
males
```

```python
#Females in each department
females = avgG[avgG["Gender"]== 'F']
females.head()
```

```python
combined_gender_df= pd.merge(females,males, how="left", on="Department_Name")
cc= combined_gender_df.fillna(0)
cc
```

```python
plt.style.use("default")
fig, ax = plt.subplots(figsize=(15,10))
ax.barh(cc.index, cc["Base_Salary_y"], label="Males", color="navy")
ax.barh(cc.index, cc["Base_Salary_x"], label="Females", color="tomato")

ax.set_xlabel("Base Salary ($)")
ax.set_ylabel("Department Name")
ax.set_title("Salary Gap Between Genders")
ax.legend()
plt.show()
```
